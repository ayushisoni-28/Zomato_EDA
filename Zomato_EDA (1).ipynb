```python
#Deleting Unecessary columns
new_data=zomato_data.drop(columns=['takeaway','opentable_support','address','url','timings','delivery','latitude','longitude','zipcode','country_id','currency'])
```

```python
new_data.isnull().sum()
```

```python
#Cusine is actually kinda important value...hence we can fill the missing values here...
new_data['cuisines']=new_data['cuisines'].fillna('No cuisine')
```

We can see in establishment the data are like [quick bites],etc which may act like noise....

```python
import numpy as np
print(new_data["establishment"].unique()[0])
new_data["establishment"] = new_data["establishment"].apply(lambda x:x[2:-2])
print(new_data["establishment"].unique()[0])

# Changing ''  to 'NA'
print(new_data["establishment"].unique())
new_data["establishment"] = new_data["establishment"].apply(lambda x : np.where(x=="", "NA", x))
print(new_data["establishment"].unique())
```

Top 10 Restaurant chains by number of outlets

```python
chains.head(10)
```

```python
top10_chains = new_data["name"].value_counts()[:10].sort_values(ascending=True)
```

Here...


data["name"].value_counts(): This counts the occurrences of each unique restaurant name in the "name" column of your DataFrame data.


[:10]: This selects the top 10 most frequent restaurant names.


.sort_values(ascending=True): This sorts the resulting Series in ascending order based on their counts.

```python
import matplotlib.pyplot as plt
height = top10_chains.values
bars = top10_chains.index
y_pos = np.arange(len(bars))

fig = plt.figure(figsize=[11,7], frameon=False)
colors = ["#f9cdac","#f2a49f","#ec7c92","#e65586","#bc438b","#933291","#692398","#551c7b","#41155e","#2d0f41"]
plt.barh(y_pos, height, color=colors)

plt.yticks(y_pos, bars, color="#424242")
plt.xlabel("Number of outlets in India")

plt.show()
```

We can see the major big food brand chains are on top 10 ...

```python
est_count = new_data.groupby("establishment").count()["res_id"].sort_values(ascending=False)[:5]

fig = plt.figure(figsize=[8,5], frameon=False)

colors = ["#2d0f41",'#933291',"#e65586","#f2a49f","#f9cdac"]
plt.bar(est_count.index, est_count.values, color=colors)

plt.xticks(range(0, 6), color="#424242")
plt.yticks(range(0, 25000, 5000), color="#424242")
plt.xlabel("Top 5 establishment types")

plt.show()
```

We can see most of the restaurents are of snack corner (quick bites) and casual dinings...

```python
city_counts = new_data.groupby("city").count()["res_id"].sort_values(ascending=True)[-10:]

height = pd.Series(city_counts.values)
bars = city_counts.index
y_pos = np.arange(len(bars))

fig = plt.figure(figsize=[11,7], frameon=False)


colors = ['#dcecc9', '#aadacc', '#78c6d0', '#48b3d3', '#3e94c0', '#3474ac', '#2a5599', '#203686', '#18216b', '#11174b']
plt.barh(y_pos, height, color=colors)

plt.xlim(3)
plt.xticks(color="#424242")
plt.yticks(y_pos, bars, color="#424242")
plt.xlabel("Number of outlets")


plt.show()
```

metro-politian cities have more restaurents compared to other cities...

```python
import seaborn as sns
sns.set_style('darkgrid')
sns.kdeplot(new_data['aggregate_rating'], shade=True)
plt.title("Ratings distribution")
plt.show()
```

On average, majority of restaurants have rating between 3 to 4 with fewer restaurants managing to go beyond 4.

```python
plt.plot("average_cost_for_two","aggregate_rating", data=new_data.sample(1000), linestyle="none", marker="o")
plt.xlim([0,3000])
plt.show()
```

This relation may concludes that that as average cost for two increases, there is a better chance that the restaurant will be rated highly.However mostly Average rating is around 3 to 4.5 in low cost range food for 2.

```python
sns.boxplot(x='price_range', y='aggregate_rating', data=new_data)
plt.ylim(1)
plt.title("Relationship between Price range and Ratings")
plt.show()
```

This graph tells us that higher the range of a food..the service may be increased resulting a good review. However we can see some outliers too representing people who may not be satisfied by the service as per the cost.
